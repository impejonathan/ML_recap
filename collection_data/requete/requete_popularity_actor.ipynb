```python
import requests
import pandas as pd

api_key = '9d100064b2baf0f1d74a7e1af0a611d8'

# # Définir la fonction pour récupérer les détails des acteurs
# def get_actor_details(actor_id, api_key):
#     url = f'https://api.themoviedb.org/3/person/{actor_id}?api_key={api_key}'
#     response = requests.get(url)

#     if response.status_code == 200:
#         actor_data = response.json()
#         return actor_data
#     else:
#         return None

# Définir la fonction pour obtenir l'actor_id par recherche d'acteur
def get_actor_id(api_key, actor_name):
    url = f'https://api.themoviedb.org/3/search/person?api_key={api_key}&query={actor_name}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        results = data.get('results', [])
        if results:
            # Récupérer l'actor_id du premier résultat (vous pouvez personnaliser la logique ici si nécessaire)
            actor_id = results[0]['id']
            return actor_id
        else:
            print(f"Acteur '{actor_name}' non trouvé dans la base de données de TMDb.")
            return None
    else:
        print(f"Erreur de requête API : {response.status_code} - {response.text}")
        return None

# Définir la fonction pour récupérer la popularité des acteurs
def get_actor_popularity(api_key, actor_name):
    actor_id = get_actor_id(api_key, actor_name)
    if actor_id:
        url = f'https://api.themoviedb.org/3/person/{actor_id}?api_key={api_key}'
        response = requests.get(url)
        if response.status_code == 200:
            actor_data = response.json()
            popularity = actor_data.get('popularity')
            return popularity
    return None


# Obtenir la popularité pour chaque acteur dans le DataFrame
df['actor_1_popularity'] = df['acteur_1'].apply(lambda actor_name: get_actor_popularity(api_key, actor_name))
df['actor_2_popularity'] = df['acteur_2'].apply(lambda actor_name: get_actor_popularity(api_key, actor_name))
df['actor_3_popularity'] = df['acteur_3'].apply(lambda actor_name: get_actor_popularity(api_key, actor_name))

# Afficher le DataFrame avec les popularités des acteurs
print(df[['acteur_1', 'actor_1_popularity', 'acteur_2', 'actor_2_popularity', 'acteur_3', 'actor_3_popularity']])
```

```python
import requests


# Définir la fonction pour rechercher les réalisateurs par leur nom et obtenir leur popularité
def get_director_popularity(api_key, director_name):
    url = f'https://api.themoviedb.org/3/search/person?api_key={api_key}&query={director_name}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        results = data.get('results', [])
        if results:
            # On suppose ici qu'il y a au plus un résultat correspondant
            return results[0].get('popularity')
    return None

# Supposez que votre DataFrame s'appelle df et que la colonne des réalisateurs s'appelle 'Realisateur'

# Obtenir la popularité de chaque réalisateur dans le DataFrame
df['director_popularity'] = df['réalisateur'].apply(lambda director_name: get_director_popularity(api_key, director_name))

# Afficher le DataFrame avec la popularité de chaque réalisateur
print(df[['réalisateur', 'director_popularity']])
```
