```python
import requests
import pandas as pd

api_key = '9d100064b2baf0f1d74a7e1af0a611d8'

# Définir la fonction pour obtenir l'ID du film par recherche de film
def get_movie_id(api_key, movie_title):
    url = f'https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={movie_title}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        results = data.get('results', [])
        if results:
            # Récupérer l'ID du film du premier résultat (vous pouvez personnaliser la logique ici si nécessaire)
            movie_id = results[0]['id']
            return movie_id
        else:
            print(f"Film '{movie_title}' non trouvé dans la base de données de TMDb.")
            return None
    else:
        print(f"Erreur de requête API : {response.status_code} - {response.text}")
        return None

# Définir la fonction pour récupérer la langue du film
def get_movie_language(api_key, movie_title):
    movie_id = get_movie_id(api_key, movie_title)
    if movie_id:
        url = f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}'
        response = requests.get(url)
        if response.status_code == 200:
            movie_data = response.json()
            language = movie_data.get('original_language')
            return language
    return None


# Supposons que vous ayez déjà importé votre DataFrame dans une variable appelée df avec une colonne 'titre' contenant les titres de films.

# Obtenir la langue pour chaque film dans le DataFrame
df['langue'] = df['titre'].apply(lambda movie_title: get_movie_language(api_key, movie_title))

# Afficher le DataFrame avec la nouvelle colonne 'langue'
print(df)
```
