```python
import pandas as pd

# Supposons que vos DataFrames s'appellent df_max et df_box_office et qu'ils contiennent tous les deux une colonne "titre"
df_merged = pd.merge(df_max, df_box_office, on="titre")

# Afficher les premières lignes du DataFrame fusionné
df_merged.head()
```

```python
df_merged
```

```python
# Suppression des colonnes 'annee_x' et 'annee_y'
df_merged = df_merged.drop(['annee_x'], axis=1)

# Réorganisation des colonnes
cols = df_merged.columns.tolist()
cols.remove('sortie_france')
cols.remove('entree')
cols = cols[:1] + ['sortie_france', 'entree'] + cols[1:]
df_merged = df_merged[cols]
```

## --------------------    ajout des estimation        ----------------------------

```python
# Ajout de la colonne '1er_semaine_estimation'
df_merged['1er_semaine_estimation'] = (df_merged['entree'] * 0.39).round().astype(int)

# Réorganisation des colonnes
cols = df_merged.columns.tolist()
cols.remove('1er_semaine_estimation')
entree_index = cols.index('entree')
cols = cols[:entree_index+1] + ['1er_semaine_estimation'] + cols[entree_index+1:]
df_merged = df_merged[cols]

df_merged
```

```python
import seaborn as sns

# Supposons que votre DataFrame s'appelle df_box_office et qu'il contient les colonnes "titre", "entree", "annee" et "pays"
top_13 = df_merged.sort_values(by="entree", ascending=False).head(20)

plt.figure(figsize=(10, 6))
sns.barplot(y="titre", x="entree", hue="pays", data=top_13)
plt.ylabel("Titre")
plt.xlabel("Nombre d'entrées")
plt.title("Les 13 films les plus vus")
plt.legend(title="Pays d'origine")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Groupby sur "franchise_1"
df_grouped = df_merged.groupby('franchise_1').sum().reset_index()

# Trier les données par "entree" en ordre décroissant et sélectionner les 10 premières lignes
df_top10 = df_grouped.sort_values('entree', ascending=False).head(13)

# Création du graphique
sns.barplot(data=df_top10, x='entree', y='franchise_1')
plt.title('Top 10 des franchises par entrées')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Groupby sur "franchise_1"
df_grouped = df_merged.groupby('franchise_2').sum().reset_index()

# Trier les données par "entree" en ordre décroissant et sélectionner les 10 premières lignes
df_top10 = df_grouped.sort_values('entree', ascending=False).head(13)

# Création du graphique
sns.barplot(data=df_top10, x='entree', y='franchise_2')
plt.title('Top 10 des franchises par entrées')
plt.show()
```

```python
# Sélectionner les lignes contenant l'un des mots spécifiés dans la colonne "titre"
selected_rows = df_merged[df_merged["titre"].str.contains(r"Les 4 Fantastiques|Coco|Robin des Bois|La Belle et la Bête|avatar")]

# Afficher les lignes sélectionnées
selected_rows
```

```python
import matplotlib.pyplot as plt

# Groupby sur "pays"
df_grouped = df_merged.groupby('pays').sum().reset_index()

# Calcul du total des entrées
total_entree = df_grouped['entree'].sum()

# Filtrage des données pour ne garder que les films ayant plus de 2% des entrées totales
df_filtered = df_grouped[df_grouped['entree'] > total_entree * 0.02]

# Création du graphique
plt.pie(df_filtered['entree'], labels=df_filtered['pays'], autopct='%1.1f%%')
plt.title('Entrées par pays (films ayant plus de 2% des entrées totales)')
plt.show()
```

```python
# Renommage de la colonne 'annee_y' en 'annee'
df_merged = df_merged.rename(columns={'annee_y': 'annee'})
```

```python
# Remplacer les valeurs NaN par une valeur par défaut (par exemple 0)
df_merged["annee"] = df_merged["annee"].fillna(0)

# Conversion des valeurs de la colonne 'annee' en nombres entiers
df_merged["annee"] = df_merged["annee"].astype(int)
```

```python
# Suppression des doublons en fonction des colonnes 'titre' et 'annee'
df_merged = df_merged.drop_duplicates(subset=['titre', 'sortie_france'])

```

```python
# df_merged.to_csv("data/gold_senscritique.csv" , index=False)   
```

```python
# Remplacement des valeurs NaN dans la colonne 'sortie_france' par les valeurs de la colonne 'annee'
df_merged['sortie_france'] = df_merged['sortie_france'].fillna(df_merged['annee'])
```

```python

import pandas as pd
import locale

# Définir la locale en français
locale.setlocale(locale.LC_TIME, 'fr_FR')

# Supposons que votre colonne "sortie_france" est actuellement de type 'object' (chaîne de caractères)
# Si ce n'est pas le cas, vous pouvez ignorer cette ligne et passer directement à la suivante.
df_merged['sortie_france'] = pd.to_datetime(df_merged['sortie_france'], format='%d %B %Y', errors='coerce')

# Maintenant, convertissons la colonne en format '28-05-2014'
df_merged['sortie_france'] = df_merged['sortie_france'].dt.strftime('%d-%m-%Y')


```
