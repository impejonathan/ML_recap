```python
def convert_to_minutes(duration):
    """Convertit une durée au format 'X h Y min' en minutes."""
    if pd.isna(duration):
        return duration
    hours = 0
    minutes = 0
    parts = duration.split()
    for i, part in enumerate(parts):
        if part == "h":
            hours = int(parts[i - 1])
        elif part == "min":
            minutes = int(parts[i - 1])
    return hours * 60 + minutes

# Convertir les durées dans la colonne "duree" en minutes
df_max["duree"] = df_max["duree"].apply(convert_to_minutes)

# Afficher les premières lignes du DataFrame résultant
print(df_max.head())
```

```python
import pandas as pd

# Création d'une fonction pour extraire les données contenant 'France' dans les colonnes 'sortie_1', 'sortie_2', 'sortie_3' et 'sortie_4'
def extract_france(row):
    for col in ['sortie_1', 'sortie_2', 'sortie_3', 'sortie_4']:
        if 'France' in str(row[col]):
            return row[col]
    return None

# Application de la fonction à chaque ligne du dataframe pour créer la nouvelle colonne 'sortie_france'
df_max['sortie_france'] = df_max.apply(extract_france, axis=1)

```

```python
# Nettoyage de la colonne 'sortie_france'
df_max['sortie_france'] = df_max['sortie_france'].str.replace('·', '').str.replace('(France)', '').str.replace('()', '').str.strip()



# Suppression des colonnes 'sortie_1', 'sortie_2', 'sortie_3' et 'sortie_4'
df_max = df_max.drop(['sortie_1', 'sortie_2', 'sortie_3', 'sortie_4'], axis=1)
```

```python
# Nettoyage de la colonne 'sortie_france'
df_max['sortie_france'] = df_max['sortie_france'].str.replace('()', '', regex=False).str.strip()

```

```python
# Suppression des doublons en fonction des colonnes 'titre' et 'annee'
df_max = df_max.drop_duplicates(subset=['titre', 'annee'])

# Affichage des premières lignes du DataFrame résultant
df_max.head()
```
