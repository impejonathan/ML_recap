```python
df = df.rename(columns={'Entrée_totales': 'box_office', 'Entrées_1ere_semaine': '1ere_semaine'})
```

```python
cols = df.columns.tolist()
cols.insert(cols.index('box_office') + 1, cols.pop(cols.index('1ere_semaine')))
df = df[cols]
```

```python
df[['box_office', 'duree', '1ere_semaine']] = df[['box_office', 'duree', '1ere_semaine']].fillna(0).astype(int)
```

```python

import pandas as pd
import locale

# Définir la locale en français
locale.setlocale(locale.LC_TIME, 'fr_FR')

# Supposons que votre colonne "sortie_france" est actuellement de type 'object' (chaîne de caractères)
# Si ce n'est pas le cas, vous pouvez ignorer cette ligne et passer directement à la suivante.
df['sortie_france'] = pd.to_datetime(df['sortie_france'], format='%d %B %Y', errors='coerce')

# Maintenant, convertissons la colonne en format '28-05-2014'
df['sortie_france'] = df['sortie_france'].dt.strftime('%d-%m-%Y')
```

```python
from datetime import datetime

vacances_scolaires = []

for year in range(2000, 2023):
    vacances_scolaires.append(("Vacances de la Toussaint", f"{year}-10-22", f"{year}-11-03"))
    vacances_scolaires.append(("Vacances de Noël", f"{year}-12-17", f"{year+1}-01-02"))
    vacances_scolaires.append(("Vacances d'hiver", f"{year+1}-02-11", f"{year+1}-02-27"))
    vacances_scolaires.append(("Vacances de printemps", f"{year+1}-04-08", f"{year+1}-04-24"))
    vacances_scolaires.append(("Grandes vacances", f"{year+1}-07-07", f"{year+1}-09-01"))

def is_vacances(date):
    if not isinstance(date, str):
        return False
    date = datetime.strptime(date, '%d-%m-%Y').date()
    for vacance in vacances_scolaires:
        start = datetime.strptime(vacance[1], '%Y-%m-%d').date()
        end = datetime.strptime(vacance[2], '%Y-%m-%d').date()
        if start <= date <= end:
            return True
    return False

df['vacances'] = df['sortie_france'].apply(is_vacances)

```
