 <font color='red'><h2>Nombre d'entrées pour les films lors de leur première semaine en fonction de leur mois de sortie</h2></font>


```python
import plotly.express as px
import pandas as pd

# Assurez-vous que la colonne 'sortie_france' est au format datetime
df['sortie_france'] = pd.to_datetime(df['sortie_france'], format='%Y-%m-%d')

# Filtrer les données pour inclure uniquement les dates entre 2015 et fin 2023
mask = (df['sortie_france'] >= '2015-01-01') & (df['sortie_france'] <= '2023-12-31')
df_filtered = df.loc[mask]

# Créer une colonne 'mois' pour regrouper les données par mois
df_filtered['mois'] = df_filtered['sortie_france'].dt.month

# Créer une colonne 'année' pour la légende
df_filtered['année'] = df_filtered['sortie_france'].dt.year

# Regrouper les données par mois et année et calculer la somme des entrées de la première semaine pour chaque mois et année
df_grouped = df_filtered.groupby(['mois', 'année'])['1ere_semaine'].sum().reset_index()

# Créer le graphique en courbes avec Plotly
fig = px.line(df_grouped, x='mois', y='1ere_semaine', color='année', title="Nombre d'entrées pour les films lors de leur première semaine en fonction de leur mois de sortie en France, par mois, de janvier à décembre, pour toutes les années entre 2015 et fin 2023")
fig.show()


```

<font color='blue'><h2>Liste des genres à conserver</h2></font>


```python
# Liste des genres à conserver
genres_to_keep = ['Drame', 'Comédie', 'Aventure', 'Thriller', 'Comédie dramatique',
       'Action', 'Arts martiaux', 'Policier', 'Science-fiction',
       'Fantastique', 'Épouvante-Horreur', 'Animation',
       'Comédie romantique', 'Nature', 'Biopic', 'Péplum', 'Société',
       'Romance', 'Comédie musicale', 'Animalier', 'Sport', 'Guerre',
       'Politique', 'Gangster', 'Catastrophe', 'Musique', 'Fantasy',
       'Documentaire', 'Western', 'Écologie','Historique','Road movie','Making-of','Sketches','Politique']

# Remplacement des valeurs
df['genre_1'] = df['genre_1'].where(df['genre_1'].isin(genres_to_keep), other='Autre')
df['genre_2'] = df['genre_2'].where(df['genre_2'].isin(genres_to_keep), other='Autre')
df['genre_3'] = df['genre_3'].where(df['genre_3'].isin(genres_to_keep), other='Autre')

```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Sélectionner les colonnes numériques de df
df_numeric = df.select_dtypes(include='number')

# Créer une copie de df pour éviter de modifier le DataFrame original
df_copy = df.copy()

# Encoder les variables catégorielles avec LabelEncoder
le = LabelEncoder()
categorical_columns = ["Titre",'acteur_1', 'acteur_2', 'acteur_3', 'realisateur', 'genre_1', 'genre_2', 'genre_3', 'franchise_1', 'franchise_2', 'pays']
for col in categorical_columns:
    df_copy[col] = le.fit_transform(df_copy[col].astype(str))

# Sélectionner les colonnes encodées de df_copy
df_encoded = df_copy[categorical_columns]

# Concaténer les colonnes numériques et les colonnes encodées
df_concat = pd.concat([df_numeric, df_encoded], axis=1)

# Créer la matrice de corrélation pour les colonnes numériques et encodées de df_concat
corr_matrix = df_concat.corr()

# Sélectionner uniquement la colonne cible '1ere_semaine'
corr_target = corr_matrix[['1ere_semaine']]

# Trier les valeurs par ordre décroissant de corrélation avec la colonne cible
corr_target = corr_target.sort_values(by='1ere_semaine', ascending=False)

# Créer la heatmap avec Seaborn
fig, ax = plt.subplots(figsize=(5, 12))
sns.heatmap(corr_target, cmap='coolwarm', annot=True, ax=ax)
plt.title("Heatmap montrant la corrélation entre les différentes colonnes de df et la colonne cible '1ere_semaine'")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Sélectionner les colonnes numériques de df
df_numeric = df.select_dtypes(include='number')

# Créer une copie de df pour éviter de modifier le DataFrame original
df_copy = df.copy()

# Encoder les variables catégorielles avec LabelEncoder
le = LabelEncoder()
categorical_columns = ["Titre",'acteur_1', 'acteur_2', 'acteur_3', 'realisateur', 'genre_1', 'genre_2', 'genre_3', 'franchise_1', 'franchise_2', 'pays','vacances']
for col in categorical_columns:
    df_copy[col] = le.fit_transform(df_copy[col].astype(str))

# Sélectionner les colonnes encodées de df_copy
df_encoded = df_copy[categorical_columns]

# Concaténer les colonnes numériques et les colonnes encodées
df_concat = pd.concat([df_numeric, df_encoded], axis=1)

# Créer la matrice de corrélation pour les colonnes numériques et encodées de df_concat
corr_matrix = df_concat.corr()

# Sélectionner uniquement la colonne cible '1ere_semaine'
corr_target = corr_matrix[['1ere_semaine']]

# Trier les valeurs par ordre décroissant de corrélation avec la colonne cible
corr_target = corr_target.sort_values(by='1ere_semaine', ascending=False)

# Afficher les scores de corrélation pour chaque variable
print(corr_target)

# Créer la heatmap avec Seaborn
fig, ax = plt.subplots(figsize=(5, 12))
sns.heatmap(corr_target, cmap='coolwarm', annot=True, ax=ax)
plt.title("Heatmap montrant la corrélation entre les différentes colonnes de df et la colonne cible '1ere_semaine'")
plt.show()
```

D’après les résultats, il semble que la colonne cible ‘1ere_semaine’ soit corrélée positivement avec les colonnes ‘box_office’ et ‘duree’. Cela signifie que plus le box-office d’un film est élevé, plus le nombre d’entrées lors de sa première semaine est élevé. De même, plus la durée d’un film est longue, plus le nombre d’entrées lors de sa première semaine est élevé.

D’autre part, la colonne cible ‘1ere_semaine’ est corrélée négativement avec les colonnes ‘Année’, ‘acteur_1’, ‘acteur_2’, ‘acteur_3’, ‘realisateur’, ‘genre_1’, ‘genre_2’, ‘genre_3’, ‘franchise_1’, ‘franchise_2’ et ‘pays’. Cela signifie que plus l’année de sortie d’un film est récente, moins le nombre d’entrées lors de sa première semaine est élevé. De même, les films ayant des acteurs, des réalisateurs, des genres ou des franchises spécifiques ont tendance à avoir un nombre d’entrées lors de leur première semaine moins élevé.
