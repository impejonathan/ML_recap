```python
import pandas as pd
```

```python
df = pd.read_csv('films_a_venir.csv')
```

```python
df['titre'] = df['titre'].str.lower()

# Liste des colonnes à supprimer
columns_to_drop = ['note_presse', 'note_spectateur', 'nominations', 'prix']

# Utilisez la méthode drop() pour supprimer les colonnes spécifiées
df = df.drop(columns=columns_to_drop)

# Utilisez la méthode str.replace() pour enlever les caractères "\n" de la colonne "date"
df['date'] = df['date'].str.replace('\n', '')
```
