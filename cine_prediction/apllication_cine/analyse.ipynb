```python
import pickle

with open('test_model_jo.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
import pandas as pd

# Créer un DataFrame avec les données fournies
data = pd.DataFrame({
    'titre': ['Farang', 'GoldenEye'],
    'acteur_1': ['Nassim Lyes', 'Pierce Brosnan'],
    'acteur_2': ['Loryn Nounay', 'Sean Bean'],
    'acteur_3': ['Olivier Gourmet', 'Izabella Scorupco'],
    'realisateur': ['Xavier Gens', 'Martin Campbell'],
    'distributeur': ['StudioCanal', 'United International Pictures (UIP)'],
    'note_presse': [3.4, 2.8],
    'duree': [99, 130],
    'genre': ['Action', 'Espionnage'],
    'pays': ['France', 'Grande-Bretagne'],
    'type': ['Long-métrage', 'Long-métrage'],
    'nominations': [0, 3],
    'prix': [0, 0],
    'annee_production': [2023, 1995]
})

# Sélectionner les colonnes à utiliser pour la prédiction
categorical_features = ["acteur_1", "acteur_2", "acteur_3", "realisateur", "distributeur", "genre", "pays"]
numerical_features = ["duree", "nominations", "prix", "annee_production"]
X = data[categorical_features + numerical_features]

# Faire la prédiction
predictions = model.predict(X)

# Afficher les prédictions
print(predictions)
```
