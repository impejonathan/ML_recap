```python
# Fonction pour compter les pays différents
def count_unique_countries(row):
    if isinstance(row, str):  # Vérifier si la valeur est une chaîne de caractères (non NaN)
        countries_list = row.split(",")  # Séparer les pays par la virgule pour obtenir une liste
        unique_countries = set([country.strip() for country in countries_list])  # Supprimer les espaces et créer un ensemble de pays uniques
        return len(unique_countries)  # Renvoyer le nombre de pays uniques
    else:
        return 0  # Renvoyer 0 pour les valeurs NaN

# Appliquer la fonction sur la colonne "pays" et créer une nouvelle colonne "nb_pays_uniques"
df['nb_pays_uniques'] = df['pays'].apply(count_unique_countries)

# Trouver la ligne qui a le plus de pays différents
max_unique_countries_row = df.loc[df['nb_pays_uniques'].idxmax()]

# Afficher les informations sur la ligne ayant le plus de pays différents
# print("Ligne avec le plus de pays différents :")
max_unique_countries_row

# # Afficher le nombre de pays différents pour cette ligne
# print("Nombre de pays différents :", max_unique_countries_row['nb_pays_uniques'])
```

```python
# Fonction pour nettoyer les dates
def clean_date(date_str):
    if isinstance(date_str, float):  # Ignorer les valeurs de type float
        return None
    date_str = date_str.strip()  # Enlever les espaces au début et à la fin
    date_str = date_str.replace(',', '') 
    date_str = date_str.replace('\n', '')  # Enlever les virgules et sauts de ligne
    
    # Trouver l'année en dehors de la première occurrence de la date
    year_pos = date_str.find('\t')
    if year_pos != -1:
        year_str = date_str[year_pos + 1:]  # Extraire l'année
        date_str = date_str[:year_pos]  # Garder seulement la première occurrence de la date
    else:
        year_str = ''
    
    # Vérifier si la date est bien formatée avec trois parties : jour, mois et année
    parts = date_str.split()
    if len(parts) != 3:
        return None  # Retourner None si la date n'est pas bien formatée
    
    day, month, year = parts  # Séparer le jour, le mois et l'année
    month_dict = {
        'janvier': '01',
        'février': '02',
        'mars': '03',
        'avril': '04',
        'mai': '05',
        'juin': '06',
        'juillet': '07',
        'août': '08',
        'septembre': '09',
        'octobre': '10',
        'novembre': '11',
        'décembre': '12'
    }
    # Formater la date avec des zéros pour le jour et le mois si nécessaire
    day = day.zfill(2)
    month = month_dict.get(month.lower(), '00')  # Utiliser '00' si le mois n'est pas valide
    return f"{day}-{month}-{year}{year_str}"
```

```python
df['date'] = df['date'].apply(clean_date)
```

```python
df['budget'] = df['budget'].replace('-', np.nan)
```

```python
def convert_budget(budget):
    if pd.isna(budget):
        return np.nan
    budget = budget.replace('US$', '').replace('$', '').replace('SEK', '').replace('mi', '000000').replace(',', '').replace('.', '').strip()
    if 'mil' in budget:
        budget = budget.replace('mil', '000')
    budget = ''.join([c for c in budget if c.isdigit()])
    if not budget:
        return np.nan
    return int(budget)

df['budget_en_$'] = df['budget'].apply(convert_budget)

```

```python
df.info()
```

```python
def update_budget(budget):
    if pd.isna(budget):
        return np.nan
    if budget > 450000000:
        return 450000000
    if budget < 1000:
        return budget * 1000
    return budget

df['budget_en_$'] = df['budget_en_$'].apply(update_budget)
```
