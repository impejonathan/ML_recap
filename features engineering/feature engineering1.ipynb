# Genres to colonnes

```python
# Séparer les genres séparés par des virgules en listes de genres
df['genres'] = df['genres'].str.split(', ')

# Compter le nombre maximum de genres pour un film
max_genre_count = df['genres'].apply(len).max()

# Créer les nouvelles colonnes pour chaque genre et les initialiser à NaN
genre_columns = [f'genre{i}' for i in range(1, max_genre_count + 1)]
for col in genre_columns:
    df[col] = float('nan')

# Remplir les nouvelles colonnes avec les genres correspondants
for idx, genres_list in enumerate(df['genres']):
    for i, genre in enumerate(genres_list):
        df.at[idx, f'genre{i+1}'] = genre


```

# Note distributeur

```python
df['moyenne_entrees_par_distributeur'] = df.groupby('distributeur')['Entrées_1ère_semaine'].transform('mean')
```

```python
# Calcul de la moyenne des entrées pour chaque distributeur
mean_entries_by_distributor = df.groupby('distributeur')['Entrées_1ère_semaine'].mean()

# Classement des distributeurs en fonction de la moyenne des entrées (du plus petit au plus grand)
ranked_distributors = mean_entries_by_distributor.rank()

# Création d'un dictionnaire de correspondance entre le classement et la note
rank_to_score = {rank: score for rank, score in enumerate(range(1, len(ranked_distributors) + 1), 1)}

# Création de la colonne 'reputation_distributeur' en utilisant le classement pour attribuer une note
df['reputation_distributeur'] = df['distributeur'].map(mean_entries_by_distributor.rank().map(rank_to_score))
```

```python
# Vous pouvez utiliser la méthode groupby de pandas pour compter le nombre de répétitions de chaque distributeur
counts = df.groupby('distributeur').size().reset_index(name='nombre_films_distributeur')

# Fusionner les résultats avec la colonne 'reputation_distributeur' pour obtenir le produit
df = df.merge(counts, on='distributeur', how='left')
# df['reputation_distributeur_1'] = df['reputation_distributeur'] * df['nombre_films_distributeur']
```
