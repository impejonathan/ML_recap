## vacances

```python
from datetime import datetime

# Dictionnaire de correspondance des noms de mois en français aux noms de mois en anglais
mois_fr_en = {
    'janvier': 'January',
    'février': 'February',
    'mars': 'March',
    'avril': 'April',
    'mai': 'May',
    'juin': 'June',
    'juillet': 'July',
    'août': 'August',
    'septembre': 'September',
    'octobre': 'October',
    'novembre': 'November',
    'décembre': 'December'
}

vacances_scolaires = []

for year in range(1200, 2025):
    vacances_scolaires.append(("Vacances de la Toussaint", f"{year}-10-22", f"{year}-11-03"))
    vacances_scolaires.append(("Vacances de Noël", f"{year-1}-12-17", f"{year}-01-02"))
    vacances_scolaires.append(("Vacances d'hiver", f"{year}-02-11", f"{year}-02-27"))
    vacances_scolaires.append(("Vacances de printemps", f"{year}-04-08", f"{year}-04-24"))
    vacances_scolaires.append(("Grandes vacances", f"{year}-07-07", f"{year}-09-01"))

def is_vacances(week):
    if not isinstance(week, str):
        return False

    # Extraire les dates de début et de fin de la semaine
    week_dates = week.split(" au ")
    if len(week_dates[0].split(' ')) == 2:
        # Cas où les dates sont de la forme 'jour1 mois1 au jour2 mois2 année'
        start_date_str = week_dates[0]
        end_date_components = week_dates[1].split(' ')  # Split the end date into components
        day, month, year = end_date_components  # Separate day, month, and year
        end_date_str = f"{day} {month} {year}"  # Join day, month, and year with spaces
        start_date_str += f" {year}" 
    elif len(week_dates[0].split(' ')) == 1 :
        # Cas où les dates sont de la forme 'jour1 au jour2 mois année'
        start_date_str = week_dates[0]
        end_date_components = week_dates[1].split(' ')  # Split the end date into components
        day, month, year = end_date_components  # Separate day, month, and year
        end_date_str = f"{day} {month} {year}"  # Join day, month, and year with spaces
        start_date_str += f" {month} {year}"  # Add the year to the start date
    elif len(week_dates[0].split(' ')) == 3:
        # Cas où les dates sont de la forme 'jour1 mois1 année1 au jour2 mois2 année2'
        start_date_str = week_dates[0]
        end_date_str = week_dates[1]
    else:
        return False

    # Convertir les noms de mois en français en noms de mois en anglais
    for mois_fr, mois_en in mois_fr_en.items():
        start_date_str = start_date_str.replace(mois_fr, mois_en)
        end_date_str = end_date_str.replace(mois_fr, mois_en)

    # Convertir les dates en objets datetime
    start_date = datetime.strptime(start_date_str, '%d %B %Y').date()
    end_date = datetime.strptime(end_date_str, '%d %B %Y').date()

    # Vérifier si la semaine correspond à une période de vacances scolaires
    for vacance in vacances_scolaires:
        start_vacance = datetime.strptime(vacance[1], '%Y-%m-%d').date()
        end_vacance = datetime.strptime(vacance[2], '%Y-%m-%d').date()
        if start_vacance <= start_date <= end_vacance or start_vacance < end_date < end_vacance:
            #print(start_date, end_date, start_vacance, end_vacance)
            return True
    #print(start_date, end_date, start_vacance, end_vacance)
    return False

#df['vacances'] = df['Semaine'].apply(is_vacances)
```

```python
df['vacances'] = df['Semaine'].apply(is_vacances)
```

## Saison

```python
from datetime import datetime

def get_season(week):
    if not isinstance(week, str):
        return False

    # Extraire les dates de début et de fin de la semaine
    week_dates = week.split(" au ")
    if len(week_dates[0].split(' ')) == 2:
        # Cas où les dates sont de la forme 'jour1 mois1 au jour2 mois2 année'
        start_date_str = week_dates[0]
        end_date_components = week_dates[1].split(' ')  # Split the end date into components
        day, month, year = end_date_components  # Separate day, month, and year
        end_date_str = f"{day} {month} {year}"  # Join day, month, and year with spaces
        start_date_str += f" {year}" 
    elif len(week_dates[0].split(' ')) == 1 :
        # Cas où les dates sont de la forme 'jour1 au jour2 mois année'
        start_date_str = week_dates[0]
        end_date_components = week_dates[1].split(' ')  # Split the end date into components
        day, month, year = end_date_components  # Separate day, month, and year
        end_date_str = f"{day} {month} {year}"  # Join day, month, and year with spaces
        start_date_str += f" {month} {year}"  # Add the year to the start date
    elif len(week_dates[0].split(' ')) == 3:
        # Cas où les dates sont de la forme 'jour1 mois1 année1 au jour2 mois2 année2'
        start_date_str = week_dates[0]
        end_date_str = week_dates[1]
    else:
        return False

    # Convertir les noms de mois en français en noms de mois en anglais
    for mois_fr, mois_en in mois_fr_en.items():
        start_date_str = start_date_str.replace(mois_fr, mois_en)
        end_date_str = end_date_str.replace(mois_fr, mois_en)

    # Convertir les dates en objets datetime
    start_date = datetime.strptime(start_date_str, '%d %B %Y').date()
    end_date = datetime.strptime(end_date_str, '%d %B %Y').date()

    # Déterminer le mois de début et de fin
    start_month = start_date.month
    end_month = end_date.month

    # Déterminer la saison correspondante en fonction du mois
    if (3 <= start_month <= 5 and 3 <= end_month <= 5) or (end_month == 6 and end_date.day <= 3) or (end_month == 3 and end_date.day > 3) :
        season = "Printemps"
    elif (6 <= start_month <= 8 and 6 <= end_month <= 8) or (end_month == 9 and end_date.day <= 3) or (end_month == 6 and end_date.day > 3):
        season = "Été"
    elif 9 <= start_month <= 11 and 9 <= end_month <= 11 or (end_month == 12 and end_date.day <= 3) or (end_month == 9 and end_date.day > 3):
        season = "Automne"
    else:
        season = "Hiver"

    return season
    

dates = [
    "20 au 27 avril 2022",
    "28 septembre au 5 octobre 2022",
    "17 au 24 août 2022",
    "07 au 14 septembre 2022",
     "28 mai au 4 juin 2022",
     '24 au 31 mai 2023'
    
]

for date in dates:
    season = get_season(date)
    print(f"{date} correspond à la saison de {season}.")
```

```python
df['saison']= df['Semaine'].apply(get_season)
```
